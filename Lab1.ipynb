```python
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

p = (a + b + c) / 2
S = math.sqrt(p * (p - a) * (p - b) * (p - c))

h_a = (2 * S) / a
h_b = (2 * S) / b
h_c = (2 * S) / c

print("   a: ", h_a)
print("   b: ", h_b)
print("   c: ", h_c)
```

## Task 3

```python
import math

x1, y1 = map(float, input(" A (x1 y1): ").split())
x2, y2 = map(float, input(" B (x2 y2): ").split())
x3, y3 = map(float, input(" C (x3 y3): ").split())

area = 0.5 * abs((x2 - x1) * (y3 - y1) - (x3 - x1) * (y2 - y1))
if area < 1e-10:
    print("    ")
else:
    ba_x, ba_y = x1 - x2, y1 - y2
    bc_x, bc_y = x3 - x2, y3 - y2
    dot = ba_x * bc_x + ba_y * bc_y
    len_ba = math.sqrt(ba_x**2 + ba_y**2)
    len_bc = math.sqrt(bc_x**2 + bc_y**2)
    cos_angle = max(-1.0, min(1.0, dot / (len_ba * len_bc)))
    angle_deg = math.degrees(math.acos(cos_angle))
    print(" ABC = ", angle_deg)
```

## Task 1

```python
x = float(input("Enter x: "))
h = float(input("Enter h: "))

x = 4.0
y_start = 2.0
y_end = 3.0
h = 0.2

print(f" : {x} ")
print(f"y  {y_start}  {y_end}    {h} \n")

y = y_start
while y <= y_end:
    if y <= x:
        cos_angle = y / x
        angle_rad = math.acos(cos_angle)
        angle_deg = math.degrees(angle_rad)
        print(f"y = {y}    = {angle_deg}")
    else:
        print(f"y = {y}    (y > x)")
    y += h
    y = round(y, 1)
```

## Task 2

```python
import math

n = int(input("Enter n: "))

sum_cos = 0.0
sum_sin = 0.0
result = 0.0

for i in range(1, n + 1):
  sum_cos += math.cos(i)
  sum_sin += math.sin(i)
  result += sum_cos / sum_sin

print(f"Result sum = {result}")
```

## Task 3

```python
eps = 0.001
n = 1
sum = 0.0

while True:
  delimeter = (3 * n - 2) * (3 * n + 1)
  a_i = 1.0 / delimeter
  if(a_i > eps):
    sum += a_i
    n += 1
  else:
    break
print(f"   eps = {eps}: {sum}")
```
