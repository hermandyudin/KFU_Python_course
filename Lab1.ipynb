## Task 1

```python
x = float(input("Enter x: "))
y = float(input("Enter y: "))
z = x * y - ( x** 3 - y ** 3) / (x ** 2 + 2 * x * y + y ** 2)

print("Result: z = ", z)
```

## Task 2

```python
import math

x = float(input("Enter x: "))
y = x ** 3 * math.log(x) + math.log(4 * math.sqrt(x) + 5)
print("Result: y = ", y)
```

## Task 3

```python
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

p = (a + b + c) / 2
S = math.sqrt(p * (p - a) * (p - b) * (p - c))

h_a = (2 * S) / a
h_b = (2 * S) / b
h_c = (2 * S) / c

print("   a: ", h_a)
print("   b: ", h_b)
print("   c: ", h_c)
```

## Task 4

```python
x = int(input("Enter x: "))

tmp = x
sum = 0
while(tmp > 0):
  sum += tmp % 10
  tmp //= 10

tmp = x
prod = 1
while(tmp > 0):
  prod *= tmp % 10
  tmp //= 10

print(" : ", sum)
print(" : ", prod)
```
